## User Interface time!

```python
import random
import time
import gradio as gr

def ai_assistant(text):
    print(f"User asked: {text}")

    responses = [
        f"Analyzing '{text}'...",
        f"Searching knowledge base for '{text}'...",
        f"Generating intelligent response for '{text}'...",
        f"Done! Here's a fun fact about '{text}'."
    ]

    time.sleep(1)

    mood = random.choice([
        "Curious",
        "Excited",
        "Focused",
        "Creative"
    ])

    score = random.randint(70, 100)

    return (
        responses[-1],
        mood,
        score,
        text[::-1]
    )

demo = gr.Interface(
    fn=ai_assistant,
    inputs=gr.Textbox(
        label="Ask Anything",
        placeholder="Type something interesting..."
    ),
    outputs=[
        gr.Textbox(label="AI Response"),
        gr.Textbox(label="AI Mood"),
        gr.Slider(0, 100, label="Confidence Score"),
        gr.Textbox(label="Reversed Input")
    ],
    title="Mini AI Assistant",
    description="A slightly more advanced Gradio demo",
    theme="soft",
    flagging_mode="never"
)

demo.launch()
```

```python
import gradio as gr

def analyze_text(text):
    words = len(text.split())
    chars = len(text)
    uppercase = text.upper()
    lowercase = text.lower()
    reversed_text = text[::-1]

    return words, chars, uppercase, lowercase, reversed_text

demo = gr.Interface(
    fn=analyze_text,
    inputs=gr.Textbox(lines=6, placeholder="Enter text..."),
    outputs=[
        gr.Number(label="Word Count"),
        gr.Number(label="Character Count"),
        gr.Textbox(label="Uppercase"),
        gr.Textbox(label="Lowercase"),
        gr.Textbox(label="Reversed")
    ],
    title="Advanced Text Analyzer",
    theme="soft",
    flagging_mode="never"
)

demo.launch()
```

## Forcing dark mode

Gradio appears in light mode or dark mode depending on the settings of the browser and computer. There is a way to force gradio to appear in dark mode, but Gradio recommends against this as it should be a user preference (particularly for accessibility reasons). But if you wish to force dark mode for your screens, below is how to do it.

```python
import gradio as gr

def shout(text):
    return text.upper()

force_light_mode = """
function () {
    const url = new URL(window.location);

    if (url.searchParams.get('__theme') !== 'light') {
        url.searchParams.set('__theme', 'light');
        window.location.href = url.href;
    }
}
"""

gr.Interface(
    fn=shout,
    inputs="textbox",
    outputs="textbox",
    js=force_light_mode,
    flagging_mode="never"
).launch()
```

```python
# Let's create a call that streams back results
# If you'd like a refresher on Generators (the "yield" keyword),
# Please take a look at the Intermediate Python guide in the guides folder
def stream_gpt(prompt):

    messages = [
        {"role": "system", "content": system_message},
        {"role": "user", "content": prompt}
    ]

    stream = client.chat.completions.create(
        model="llama3.2:latest",
        messages=messages,
        stream=True
    )

    result = ""

    for chunk in stream:

        content = chunk.choices[0].delta.content

        if content:
            result += content
            yield result
```

```python
view = gr.Interface(
    fn=stream_gpt,
    title="Local Llama Chat",
    description="Streaming AI responses with Ollama + Gradio",
    theme="soft",
    inputs=gr.Textbox(
        label="Your Message",
        placeholder="Ask anything...",
        lines=7
    ),
    outputs=gr.Markdown(label="Response"),
    examples=[
        ["Explain transformers simply"],
        ["Teach me binary search"],
        ["How do LLMs work internally?"]
    ],
    flagging_mode="never"
)

view.launch()
```

```python
def stream_claude(prompt):
    messages = [
        {"role": "system", "content": system_message},
        {"role": "user", "content": prompt}
      ]
    stream = anthropic.chat.completions.create(
        model='claude-sonnet-4-5-20250929',
        messages=messages,
        stream=True
    )
    result = ""
    for chunk in stream:
        result += chunk.choices[0].delta.content or ""
        yield result
```

Remember to check the Intermediate Python Guide if you're unsure about generators and "yield"

```python
def stream_model(prompt, model):

    if model == "Llama 3.2":
        result = stream_llama(prompt)

    elif model == "Gemma":
        result = stream_gemma(prompt)

    else:
        raise ValueError("Unknown model")

    yield from result
```

```python
message_input = gr.Textbox(
    label="Your message:",
    info="Enter a message for the LLM",
    lines=7
)

model_selector = gr.Dropdown(
    ["Llama 3.2", "Gemma"],
    label="Select model",
    value="Llama 3.2"
)

message_output = gr.Markdown(label="Response:")

view = gr.Interface(
    fn=stream_model,
    title="Local LLM Playground",
    description="Chat with multiple local models using Ollama",
    theme="soft",
    inputs=[message_input, model_selector],
    outputs=[message_output],
    examples=[
        ["Explain the Transformer architecture simply", "Llama 3.2"],
        ["Teach me recursion with examples", "Gemma"]
    ],
    flagging_mode="never"
)

view.launch()
```
